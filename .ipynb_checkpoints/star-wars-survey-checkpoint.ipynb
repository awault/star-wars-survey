# Star Wars: Fan Favorites

```python
# Import the dataset

import pandas as pd
star_wars = pd.read_csv("../Data/star_wars.csv", encoding="ISO-8859-1")
```

```python
# Review columns in the dataset

star_wars.columns
```

```python
# Preview the dataframe

star_wars
```

```python
# Create a dictionary to map Y or N answers with True or False

yes_no = {
    "Yes": True,
    "No": False
}
```

```python
# Map survey responses with Boolean values

viewed_1_or_more = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no) 
fan_of_franchise = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
# Review Value Counts

viewed_1_or_more.value_counts()
```

```python
# Review Value Counts

fan_of_franchise.value_counts()
```

```python
# Create Movie Map for Y/N Responses

import numpy as np

movie_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}


# Use a loop to appy the movie map to the applicable columns

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_map)
    
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```
