# Star Wars: Fan Favorites

```python
# Import the dataset

import pandas as pd
star_wars = pd.read_csv("../Data/star_wars.csv", encoding="ISO-8859-1")
```

```python
# Review columns in the dataset

star_wars.columns
```

```python
# Preview the dataframe

star_wars
```

```python
# Create a dictionary to map Y or N answers with True or False

yes_no = {
    "Yes": True,
    "No": False
}
```

```python
# Map survey responses with Boolean values and save in new column

star_wars["viewed_1_or_more"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no) 
star_wars["fan_of_franchise"]= star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
# Review Value Counts

star_wars["viewed_1_or_more"].value_counts()
```

```python
# Review Value Counts

star_wars["fan_of_franchise"].value_counts()
```

```python
# Create Movie Map for Y/N Responses to columns [3:9]

import numpy as np

movie_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

# Use a loop to appy the movie map to the applicable columns

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_map)
```

```python
# Rename columns [3:9]

star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars.iloc[:,9:15]
```

```python
# Rename columns [9:15] to evaluate episode ranks

star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })
```

```python
# Display first 5 of columns [9:15]

star_wars.iloc[0:5,9:15]
```

```python
# Convert rankings columns to float [9:15]

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# Recalibrate rankings to 6 Best / 1 Least Favorite

def recalc_rank(rank):
    new_rank = 7 - rank
    rank = new_rank
    return rank

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].apply(recalc_rank)
star_wars.iloc[0:5,9:15]
```

## Episode Rankings

```python
# Calculate average ranking for each episode (Higher is Better)

episode_rank = star_wars[star_wars.columns[9:15]].mean()
print(episode_rank)
```

```python
import matplotlib.pyplot as plt

x_label = star_wars.columns[9:15]
plt.bar(x_label, episode_rank)
plt.xticks(rotation=30,fontsize=10)
plt.xlabel('Star Wars Episode',fontsize=14)
plt.ylabel('Average Rank',fontsize=14)
plt.show()

```

Survey respondents were asked to rank their favorite episodes of Star Wars. We have re-calibrated the survey rankings, so the highest value corresponds to the highest ranking. (e.g. Best Episode = 6, Worst Episode = 1) The bar chart above displays the average ranking for each episode. 

**Star Wars: Episode V The Empire Strikes Back, is clearly the fan favorite.**

## View Counts

```python
# Calculate the total number of survey respondents that have viewed each episode

view_count = star_wars[star_wars.columns[3:9]].sum()
view_count
```

```python
x_label = star_wars.columns[3:9]
plt.bar(x_label,view_count)
plt.xticks(rotation=30,fontsize=10)
plt.xlabel('Star Wars Episode',fontsize=14)
plt.ylabel('Total Viewed',fontsize=14)
plt.show()
```

The original episodes of Star Wars, in particular episodes 5 and 6, were seen by more of the survey respondents. This reinforces the results from the ranking data. The "original" episodes of Star Wars appear to be more popular.
